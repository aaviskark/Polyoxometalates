```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = pd.read_excel('alphaPOM.xlsx') #finds excel table and sets it = to x

x.head()

xval = x['cm**-1'] #sets X value = to column titled 'cm**1'
yval = x['Intensity'] #sets Y value = to colum titled 'Intensity'

plt.plot(xval, yval/max(yval), label = 'Exp') # plots x value; yval is plot with data normalized to max value
plt.vlines([733.179270,728.519253,702.218327, 674.599548,653.855551,625.190288,611.089748,602.409329], min(yval) -.25, 1, color = 'red', label = 'DFT') # plots DFT values and sets min to minimum yval - .1
plt.hlines(0,0,max(xval), color = 'gray', linestyle = 'dashed') # plots a dashed line at (0,0)
plt.title('Alpha POM IR Spectrum', size = 18) 
plt.xlabel("Frequency (cm^-1)",size = 15)
plt.ylabel("Relative Absorbance", size = 15)
plt.xlim(1200, min(xval)) # inverts X axis to go from highest to lowest
plt.ylim(min(yval) -.25, 1.35) # sets y axis min to start at minimun yval -.1; max at 1.30 
plt.yticks([])
plt.legend(loc = 'upper right')

plt.tight_layout()
plt.savefig('./IR_alphaPOM.pdf')
plt.show()
```
